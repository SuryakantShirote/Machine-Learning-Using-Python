**Removing Outliers**

```python
for col in df.columns:
 Q1=df[col].quantile(0.25)
 Q3=df[col].quantile(0.75)
 IQR= Q3-Q1
 lower_limit= Q1 - 1.5 *IQR
 upper_limit = Q3 + 1.5 * IQR
 df= df[(df[col]>=lower_limit) & (df[col]<=upper_limit)]
```

**Using VIF , to remove 'Multicollinearity'**

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
col_list =[]
for col in df.columns:
    if (df[col].dtypes != 'object') &  (col != 'target'):
        col_list.append(col)

X = df[col_list]
vif_data = pd.DataFrame()
vif_data['features'] = X.columns
vif_data['vif'] = [variance_inflation_factor(X.values,i)for i in range(len(X.columns))]
print(vif_data)
```

**Assigning dependant and independant variable**

```python
x = df.drop('target',axis=1)
y = df.target
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =  train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import RobustScaler
rs = RobustScaler()
```

```python
x_train = rs.fit_transform(x_train)
x_test  = rs.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
pred = lr.predict(x_test)
```

```python
from sklearn.metrics import *
```

```python
lr_accuracy = accuracy_score(y_test,pred)
print("Accuracy of Logistic regression model is ",lr_accuracy)
```

```python
lcm = confusion_matrix(y_test,pred)
lcm
```

```python
print(classification_report(y_test,pred))
```

```python
print('F1 score:', f1_score(y_test, pred,average='weighted'))
print('Recall:', recall_score(y_test,pred,
                              average='weighted'))
print('Precision:', precision_score(y_test, pred,
                                    average='weighted'))
```

```python
#lcm = confusion_matrix(y_test, y_pred)
cm_plot = sns.heatmap(lcm, annot=True, cmap='Blues');
cm_plot.set_xlabel('Predicted Values')
cm_plot.set_ylabel('Actual Values')
cm_plot.set_title('Confusion Matrix', size=16)
```
